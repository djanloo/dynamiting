```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
N = 1000
dim = 4
X = np.random.normal(0,.1, size=N*dim).reshape(-1, dim)
Y = np.random.normal(0,.2, size=N*dim).reshape(-1, dim) + np.array([2,5,2,3])
W = np.random.normal(0,.2, size=N*dim).reshape(-1, dim) + np.array([0,1,0,0])

Z = np.vstack((X,Y,W))
label = np.zeros(3*N)
label[N:2*N] = 1
label[2*N:] = 2
print(Z.shape)
sns.histplot(data=Z)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, QuantileTransformer
import umap 

scaler = QuantileTransformer()
Z = scaler.fit_transform(Z)
reducer = umap.UMAP(n_neighbors=50)
embedding = reducer.fit_transform(Z)
```

```python
sns.scatterplot(x=embedding[:,0], y=embedding[:,1], hue=label)
```
